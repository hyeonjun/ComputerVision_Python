```python
print(img.shape) # R, G, B, alpha() => 4 
type(img)
```

```python
img[0, 0]
```

```python
plt.axis('off')
plt.imshow(img)
```

```python
r, g, b = img[:,:,0], img[:,:,1], img[:,:,2]
rb = np.pad(array=r, pad_width=10, mode='constant', constant_values=1)
gb = np.pad(array=g, pad_width=10, mode='constant', constant_values=0)
bb = np.pad(array=b, pad_width=10, mode='constant', constant_values=0)
img_b = np.dstack(tup=(rb, gb, bb))
plt.axis('off')
plt.imshow(img_b)
```
