#### 행렬의 생성의 초기화
    np.empty((행, 열), 타입[np.uint8 or np.float64])
    np.zeros((행, 열), 타입)
    np.ones((행, 열), 타입)
    np.full((행, 열), 채울 값, 타입)
    np.array((행, 열), 타입)

```python
img1 = np.empty((480, 640), np.uint8)
img2 = np.zeros((480, 640), np.uint8)
img3 = np.ones((480, 640), np.uint8)
img4 = np.full((480, 640), 0, np.float64)
img5 = [[4*i+j for j in range(1, 5)] for i in range(3)]
img5 = np.array(img5)
```

#### 행렬 원소값 참조

```python
np.arange(12)
```

```python
mat1 = np.arange(12).reshape(3, -1)
mat2 = np.zeros(mat1.shape, type(mat1))

h, w = mat1.shape
for i in range(h):
    for j in range(w):
        mat2[i][j] = mat1[i][j] + 10

print(mat1)
print(mat2)
```

```python
mat1 = np.arange(12).reshape(3, -1)

mat2 = mat1 + 10 # broadcasting
print(mat1)
print(mat2)
```

### 이미지에 테두리 그리기

```python
img = plt.imread('/content/drive/MyDrive/Encore/ComputerVision/data/psj.png')
plt.axis('off')
plt.imshow(img)
```

```python
print(img.shape) # R, G, B, alpha(투명도) => 4차원 
type(img)
```

```python
img[0, 0, 0] = 1 # R에 최대 밝기값 부여
img[0, 0, 1] = 0 # G에 최소 밝기값 부여
img[0, 0, 2] = 0 # B에 최소 밝기값 부여
```

```python
img[0, 0]
```

```python
plt.axis('off')
plt.imshow(img)
```

```python
r, g, b = img[:,:,0], img[:,:,1], img[:,:,2]
rb = np.pad(array=r, pad_width=10, mode='constant', constant_values=1)
gb = np.pad(array=g, pad_width=10, mode='constant', constant_values=0)
bb = np.pad(array=b, pad_width=10, mode='constant', constant_values=0)
img_b = np.dstack(tup=(rb, gb, bb))
plt.axis('off')
plt.imshow(img_b)
```

```python
img = plt.imread('/content/drive/MyDrive/Encore/ComputerVision/data/psj.png')

# # 위 가로
# img[0:10, :, 0] = 0 # R
# img[0:10, :, 1] = 0 # G
# img[0:10, :, 2] = 1 # B

# # 아래 가로
# img[-10:, :, 0] = 0 # R
# img[-10:, :, 1] = 0 # G
# img[-10:, :, 2] = 1 # B

# # 왼쪽 세로
# img[:, :10, 0] = 0 # R
# img[:, :10, 1] = 0 # G
# img[:, :10, 2] = 1 # B

# # 오른쪽 세로
# img[:, -10:, 0] = 0 # R
# img[:, -10:, 1] = 0 # G
# img[:, -10:, 2] = 1 # B

# 정리
# 위 가로
img[:10, :] = [0, 0, 1, 1]
# 아래 가로
img[-10:, :] = [0, 0, 1, 1]
# 왼쪽 세로
img[:, :10] = [0, 0, 1, 1]
# 오른쪽 세로
img[:, -10:] = [0, 0, 1, 1]


plt.imshow(img)
```
